```python
from melfa_kinematics.transforms import fk_frames
from melfa_kinematics.plotting import plot_chain

dh_params = [
    (0, np.pi / 2, 300, 0),
    (250, 0, 0, np.pi / 6),
    (160, 0, 0, -np.pi / 8),
]

frames = fk_frames(dh_params)

plot_chain(frames)
```

```python
from melfa_kinematics.transforms import dh
from melfa_kinematics.plotting import plot_chain

T0 = np.eye(4)

T1 = dh(0, np.pi / 2, 300, 0)
T2 = T1 @ dh(250, 0, 0, np.pi / 6)
T3 = T2 @ dh(160, 0, 0, -np.pi / 8)

plot_chain([T0, T1, T2, T3])
```
